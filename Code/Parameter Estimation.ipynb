```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
from matplotlib import pylab
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

# Likelihood

```python
def likelihood(counts, probs):
    return np.product([np.power(probs[i],counts[i]) for i in range(len(counts))])
```

```python
eps=0
thetas = np.linspace(eps, 1-eps, num=10000)
```

```python
#plt.plot(thetas, [likelihood([1, 4], [t, 1-t]) for t in thetas], label="[1, 4]")
plt.plot(thetas, [likelihood([2, 3], [t, 1-t]) for t in thetas], label="[2, 3]")
plt.plot(thetas, [likelihood([3, 2], [t, 1-t]) for t in thetas], label="[3, 2]")
#plt.plot(thetas, [likelihood([4, 1], [t, 1-t]) for t in thetas], label="[4, 1]")
plt.legend()
```

# Beta Distribution

```python
a=60000
b=40000
plt.plot(thetas, beta.pdf(thetas, a, b), label="Beta["+str(a)+", "+str(b)+"]")
plt.legend()
```

```python
betas = beta.pdf(thetas, a, b)
print("Estimated mode:\t %0.4f" %thetas[np.argmax(betas)])
print("Actual mode:\t %0.4f" %((a-1)/(a+b-2)))
print("Estimated mean:\t %0.4f" %np.mean(thetas*betas))
print("Actual mean:\t %0.4f" %((a)/(a+b)))
```

# Uniform Prior

```python
prior = np.asarray([1, 1])
plt.plot(thetas, beta.pdf(thetas, prior[0], prior[1]), label="prior~Beta[%d, %d]" %(prior[0], prior[1]))

data = np.asarray([2, 3])
plt.plot(thetas, beta.pdf(thetas, prior[0]+data[0], prior[1]+data[1]), label="posterior~Beta[%d+%d, %d+%d]" %(prior[0], data[0], prior[1], data[1]))

data += np.asarray([2, 3])
plt.plot(thetas, beta.pdf(thetas, prior[0]+data[0], prior[1]+data[1]), label="posterior~Beta[%d+%d, %d+%d]" %(prior[0], data[0], prior[1], data[1]))

data += np.asarray([2, 3])
plt.plot(thetas, beta.pdf(thetas, prior[0]+data[0], prior[1]+data[1]), label="posterior~Beta[%d+%d, %d+%d]" %(prior[0], data[0], prior[1], data[1]))

plt.legend()
```
